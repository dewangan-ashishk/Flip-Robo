# 2.  In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
product_link=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')
product_url=[]

for i in product_link:
    product_url.append(i.get_attribute('href'))
    
for i in range(0,2):
    next_button=driver.find_element_by_xpath('//li[@class="a-last"]')
    try:
        next_button.click()
        
    except NoSuchElementException:
        pass
    time.sleep(5)
    new_product_link=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')
    
    for i in new_product_link:
        product_url.append(i.get_attribute('href'))
        
```

```python
brand=[]
product_name=[]
rating=[]
number_of_rating=[]
price=[]
return_exchange=[]
expected_delivery=[]
availability=[]
other_details=[]

for i in product_url:
    driver.get(i)
    time.sleep(2)
    
    try:
        brand_tag=driver.find_element_by_xpath('//td[@class="a-span9"]')
        brand.append(brand_tag.text)
    
    except NoSuchElementException:
        brand.append('-')
    
    prodname=driver.find_element_by_xpath('//h1[@id="title"]')
    product_name.append(prodname.text)
    
    try:
        rating_tag=driver.find_element_by_id('acrPopover').get_attribute('title')
        rating.append(rating_tag)
    except NoSuchElementException:
        rating.append('-')
        
    try:
        number=driver.find_element_by_id('acrCustomerReviewLink')
        number_of_rating.append(number.text)
    
    except NoSuchElementException:
        number_of_rating.append('-')
        
    try:
        price_tag=driver.find_element_by_xpath('//td[@class="a-span12"]')
        price.append(price_tag.text)
    
    except NoSuchElementException:
        price.append('-')
    ret_exchng=driver.find_element_by_xpath('//div[@class="a-section a-spacing-none icon-content"]')
    return_exchange.append(ret_exchng.text)
    
    delivery=driver.find_element_by_id('dynamicDeliveryMessage')
    expected_delivery.append(delivery.text)
    
    availability_tag=driver.find_element_by_id('availability')
    availability.append(availability_tag.text)
    
    details=driver.find_element_by_id('feature-bullets')
    other_details.append(details.text)
    
```

```python
Data=pd.DataFrame({'Brand':(brand),
                   'Product Name':(product_name),
                  'Rating':(rating),
                  'Number of Rating':(number_of_rating),
                  'Price':price,
                  'Return/Exchange':(return_exchange),
                  'Expected Delivery':(expected_delivery),
                  'Availability':(availability),
                  'Other Details':(other_details)})

Data
```

# 3.  Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
import selenium
from selenium import webdriver
import os
import time
import urllib.request
from selenium.webdriver.common.keys import Keys
```

```python
term=['fruits','cars','machine learning']
```

```python
for i in term:
    driver = webdriver.Chrome('chromedriver.exe')
    driver.get('https://www.google.com/')
    search = driver.find_element_by_name('q')
    search.send_keys(i,Keys.ENTER)
    
    time.sleep(2)
    
    img_button = driver.find_element_by_link_text('Images')
    img_button.get_attribute('href')
    img_button.click()
    
    time.sleep(2)
    
    for _ in range(200):
        driver.execute_script("window.scrollBy(0,1000)")
    
    time.sleep(2)
    image = driver.find_elements_by_xpath('//img[contains(@class,"rg_i")]')
    
    count=0
    for x in image:
        src = x.get_attribute('src')
        try:
            if src != None:
                src  = str(src)
                count+=1
                urllib.request.urlretrieve(src, os.path.join('images/'+i,i+'_image'+str(count)+'.jpg'))
            else:
                raise TypeError
        except TypeError:
            pass
    driver.close()
```

# 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1stpage. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”,“Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.flipkart.com/')
```

```python
#manually close the pop up for log in
smartphone=input('Enter the Smartphone to be searched: ')
```

```python
search_tab=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_tab.send_keys(smartphone)

search_button=driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_button.click()
```

```python
product_link=driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')
product_url=[]

for i in product_link:
    product_url.append(i.get_attribute('href'))
```

```python
brand=[]
smartphone=[]
colour=[]
RAM=[]
storage=[]
primarycam=[]
secondarycam=[]
display_size=[]
display_res=[]
processor=[]
processor_cores=[]
battery_capacity=[]
price=[]

for i in product_url:
    driver.get(i)
    time.sleep(3)
    
    brand_tag=driver.find_element_by_xpath('//h1[@class="yhB1nd"]')
    a=brand_tag.text.find(' ')
    brand.append(brand_tag.text[:a])
    
    price_tag=driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
    price.append(price_tag.text)
    readmore_button=driver.find_element_by_xpath('// button[@class="_2KpZ6l _1FH0tX"]')
    readmore_button.click()
    
    info=driver.find_elements_by_xpath('//tr[@class="_1s_Smc row"]')
    for x in info:
        str=x.text
        if (str.find('Model Name')==0):
            smartphone.append(str.replace('Model Name\n',''))
        elif (str.find('Color')==0):
            colour.append(str.replace('Color\n',''))
        elif (str.find('Internal Storage')==0):
            storage.append(str.replace('Internal Storage\n',''))
        elif (str.find('RAM')==0):
            RAM.append(str.replace('RAM\n',''))
        elif (str.find('Primary Camera\n')==0):
            primarycam.append(str.replace('Primary Camera\n',''))
        elif (str.find('Secondary Camera\n')==0):
            secondarycam.append(str.replace('Secondary Camera\n',''))
        elif (str.find('Display Size')==0):
            display_size.append(str.replace('Display Size\n',''))
        elif (str.find('Resolution\n')==0):
            display_res.append(str.replace('Resolution\n',''))
        elif (str.find('Processor Type')==0):
            processor.append(str.replace('Processor Type\n',''))
        elif (str.find('Processor Core')==0):
            processor_cores.append(str.replace('Processor Core\n',''))
        elif (str.find('Battery Capacity')==0):
            battery_capacity.append(str.replace('Battery Capacity\n',''))
```

```python
Data=pd.DataFrame({'Brand Name':(brand),
                  'Smartphone Name':(smartphone),
                  'Colour':(colour),
                  'RAM':(RAM),
                  'Storage':(storage),
                  'Primary Camera':(primarycam),
                  'Secondary Camera':(secondarycam),
                  'Display Size':(display_size),
                  'Display Resolution':(display_res),
                  'Processor':(processor),
                  'Processor Cores':(processor_cores),
                  'Battery Capacity':(battery_capacity),
                  'Price':(price),
                  'Product Url':(product_url)})
Data
```

# 10.  Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.  

```python
import selenium
from selenium import webdriver
import time
import pandas as pd

driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.hostelworld.com/')
```

```python
search=driver.find_element_by_xpath('//input[@class="location-text"]')
search.clear()
search.send_keys('London')
#execute this code twice
```

```python
driver.find_element_by_xpath('//div[@class="label"]').click()
```

```python
go=driver.find_element_by_xpath('//button[@id="search-button"]')
go.click()
```

```python
hostel=driver.find_elements_by_xpath('//h2[@class="title title-6"]//a')
dist=driver.find_elements_by_xpath('//div[@class="subtitle body-3"]')
ratings=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
#total=driver.find_elements_by_xpath('//div[@class="reviews"]')
overall=driver.find_elements_by_xpath('//div[@class="keyword"]')
price=driver.find_elements_by_xpath('//div[@class="price-col"]')

name=[]
distance=[]
rating=[]
total_review=[]
overall_review=[]
private=[]
dorm=[]
hostel_url=[]

for i in range(0,len(hostel)):
    name.append(hostel[i].text)
    distance.append(dist[i+3].text)
    x=ratings[i].text.split()
    rating.append(x[0])
    total_review.append(x[-3])
    overall_review.append(overall[i].text)
    private.append(price[2*i].text)
    dorm.append(price[2*i+1].text)
    hostel_url.append(hostel[i].get_attribute('href'))
```

```python
for i in range(2):
    next_button=driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')
    next_button.click()
    hostel=driver.find_elements_by_xpath('//h2[@class="title title-6"]//a')
    dist=driver.find_elements_by_xpath('//div[@class="subtitle body-3"]')
    ratings=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
    #total=driver.find_elements_by_xpath('//div[@class="reviews"]')
    overall=driver.find_elements_by_xpath('//div[@class="keyword"]')
    price=driver.find_elements_by_xpath('//div[@class="price-col"]')
    
    for i in range(0,len(hostel)):
        name.append(hostel[i].text)
        distance.append(dist[i].text)
        x=ratings[i].text.split()
        rating.append(x[0])
        total_review.append(x[-3])
        overall_review.append(overall[i].text)
        private.append(price[2*i].text)
        dorm.append(price[2*i+1].text)
        hostel_url.append(hostel[i].get_attribute('href'))
```

```python
description=[]
facilities=[]
for i in hostel_url:
    driver.get(i)
    try:
        more=driver.find_element_by_xpath('//a[@class="toggle-content"]')
        more.click()
    except:
        pass
    desc=driver.find_element_by_xpath('//div[@class="collapsible"]')
    description.append(desc.text)
    facility=driver.find_elements_by_xpath('//ul[@class="facilities"]')
    fac=[]
    for x in facility:
        fac.append(x.text)
    facilities.append(fac)
```

```python
Data=pd.DataFrame({'Hostel Name':(name),
                  'Distance from City Centre':(distance),
                  'Rating':(rating),
                  'Total Reviews':(total_review),
                  'Overall revies':(overall_review),
                  'Starting Price of Private Room':(private),
                  'Starting Price of Dorm':(dorm),
                  'Property Description':(description),
                  'Facilities':(facilities)})
Data
```
