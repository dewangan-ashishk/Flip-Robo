```python

title_tag=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
location_tag=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]//span')
company_tag=driver.find_elements_by_xpath('//div[@class="mt-7 companyInfo subheading lh16"]')
experience_tag=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]//span')

title=[]
location=[]
company=[]
experience=[]

for i in range(0,10):
    title.append(title_tag[i].text)
    location.append(location_tag[i].text)
    company.append(company_tag[i].text)
    experience.append(experience_tag[i].text)


Data=pd.DataFrame({'Job Title':(title),
                  'Location':(location),
                  'Company Name':(company),
                  'Experience Required':(experience)})

Data
```

# Q2: Write a python program to scrape data for “Data Scientist” Job position in  “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd

driver=webdriver.Chrome('chromedriver.exe')
url='https://naukri.com'
driver.get(url)

search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Analyst')

search_location=driver.find_element_by_name('location')
search_location.send_keys('Bangalore')

search_button=driver.find_element_by_xpath('//button[@class="btn"]')
search_button.click()
```

```python
title_tag=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
location_tag=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]//span')
company_tag=driver.find_elements_by_xpath('//div[@class="mt-7 companyInfo subheading lh16"]')
experience_tag=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]//span')

title=[]
location=[]
company=[]
job_url=[]
jd=[]

for i in range(0,10):
    title.append(title_tag[i].text)
    location.append(location_tag[i].text)
    company.append(company_tag[i].text)
    
url_tag=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')

for i in url_tag:
    job_url.append(i.get_attribute('href'))

for i in job_url[:10]:
    try:
        driver.get(i)
        tag_desc=driver.find_element_by_xpath('//section[@class="job-desc"]')
        jd.append(tag_desc.text)
    except NoSuchElementException:
        jd.append('-')
        

Data=pd.DataFrame({'Job Title':(title),
                  'Location':(location),
                  'Company Name':(company),
                  'Job Description':(jd)})

Data
```

# Q8: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field. You have to scrape 4 attributes of each sneaker : 1. Brand 2. Product Description 3. Price 4. discount %

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.flipkart.com/')
```

```python
search_tab=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_tab.send_keys('Sneakers')

search_button=driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_button.click()
```

```python
urls=driver.find_elements_by_xpath('//a[@class="_2UzuFa"]')
link=[]
for i in urls:
    link.append(i.get_attribute('href'))
for i in range(0,2):
    nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")

    try:

        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page

    except:

        driver.get(nxt_button[0].get_attribute('href'))
    next_url=driver.find_elements_by_xpath('//a[@class="_2UzuFa"]')
    time.sleep(5)
    for x in next_url:
        link.append(x.get_attribute('href'))
```

```python
Brand=[]
PD=[]
Price=[]
Discount=[]
for i in link[:100]:
    driver.get(i)
    try:
        brand_tag=driver.find_element_by_xpath('//span[@class="G6XhRU"]')
        Brand.append(brand_tag.text)
    except NoSuchElementException:
        Brand.append('-')
    
    try:
        pd_tag=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        PD.append(pd_tag.text)
    except NoSuchElementException:
        PD.append('-')
    
    try:
        price_tag=driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        Price.append(price_tag.text)
    except NoSuchElementException:
        Price.append('-')
        
    try:
        discount_tag=driver.find_element_by_xpath('//div[@class="_3Ay6Sb _31Dcoz pZkvcx"]//span')
        Discount.append(discount_tag.text)
    except NoSuchElementException:
        Discount.append('0 % off')
    
    
        
```

```python
Data=pd.DataFrame({'Brand':(Brand),
                  'Product Description':(PD),
                  'Price':(Price),
                  'Discount':(Discount)})

Data
```

# Q9: Go to the link - https://www.myntra.com/shoes Set Price filter to “Rs. 6649 to Rs. 13099” , Color filter to “Black”. And then scrape First 100 shoes data you get. The data should include “Brand” of the shoes , Short Shoe description, price of the shoe

```python
import selenium
from selenium import webdriver
import pandas as pd
import time

driver=webdriver.Chrome('chromedriver.exe')
url='https://www.myntra.com/shoes'
driver.get(url)
```

```python
pricefilter_tag=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label')
pricefilter_tag.click()
```

```python
colorfilter_tag=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label')
colorfilter_tag.click()
```

```python
brand_tag=driver.find_elements_by_xpath('//h3[@class="product-brand"]')
description_tag=driver.find_elements_by_xpath('//h4[@class="product-product"]')
price_tag=driver.find_elements_by_xpath('//div[@class="product-price"]')
       
brand=[]
description=[]

for i in range(len(brand_tag)):
    brand.append(brand_tag[i].text)
    description.append(description_tag[i].text)
    
price=[]
for j in price_tag:
    str=j.text
    if (len(str)>=10):
        for i in range(0,len(str)):
            if (str[i].isnumeric() & str[i+1].isalpha()):
                price.append(str[:i+1])
                break
    else:
        price.append(j.text)
```

```python
while (len(brand)<100):
    nxt_button=driver.find_elements_by_xpath("//a[@rel='next']")

    try:

        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page

    except:

        driver.get(nxt_button[0].get_attribute('href'))
    
    time.sleep(5)
    
    brand_tag=driver.find_elements_by_xpath('//h3[@class="product-brand"]')
    description_tag=driver.find_elements_by_xpath('//h4[@class="product-product"]')
    price_tag=driver.find_elements_by_xpath('//div[@class="product-price"]')
    for j in price_tag:
        str=j.text
        if (len(str)>=10):
            for i in range(0,len(str)):
                if (str[i].isnumeric() & str[i+1].isalpha()):
                    price.append(str[:i+1])
                    break
        else:
            price.append(j.text)
    
    for i in range(len(brand_tag)):
        brand.append(brand_tag[i].text)
        description.append(description_tag[i].text)
        
        
```

```python
Data=pd.DataFrame({'Brand':(brand),
                  'Product Description':(description),
                  'Price':(price)})
Data
```
