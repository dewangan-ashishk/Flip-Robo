```python
rating=[]
review=[]
full_review=[]
```

```python
for url in link:
    driver.get(url)
    
    last_lim=len(rating)
    
    rating_tag=driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]')
    review_tag=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
    full_review_tag=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')

    for i in range(0,len(rating_tag)):
        str=rating_tag[i].text
        if str!='':
            rating.append(int(str[0]))
        else:
            rating.append(None)
        review.append(review_tag[i].text)
        full_review.append(full_review_tag[i].text)
        
    try:
        all_review=driver.find_elements_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[8]/div/a')
        review_count=driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]')
        lim = int(''.join(n for n in review_count.text if n.isnumeric()))
        
        # Setting a threshold for number of reviews to be extracted for a product
        if lim>500:
            lim=500
        
        if len(all_review)==0:
            all_reviews=driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]')
            all_reviews.click()
            
            time.sleep(2)
            rating_tag=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]')
            review_tag=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
            full_review_tag=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    
            for i in range(0,len(rating_tag)):
                str=rating_tag[i].text
                if str!='':
                    rating.append(int(str[0]))
                else:
                    rating.append(None)
                review.append(review_tag[i].text)
                full_review.append(full_review_tag[i].text)
    
            while (len(rating)<last_lim+lim):
                nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
                if len(nxt_button)!=0:
                    try:
                        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page
            
                    except NoSuchElementException:
                        pass
            
                    except:
                        driver.get(nxt_button[0].get_attribute('href'))
    
                time.sleep(2)
                rating_tag=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]')
                review_tag=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
                full_review_tag=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    
                for i in range(0,len(rating_tag)):
                    str=rating_tag[i].text
                    if str!='':
                        rating.append(int(str[0]))
                    else:
                        rating.append(None)
                    review.append(review_tag[i].text)
                    full_review.append(full_review_tag[i].text)
            
            
        else:
            review_page=all_review[0].get_attribute('href')
            driver.get(review_page)

            while (len(rating)<last_lim+lim):
                nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
                if len(nxt_button)!=0:
                    try:
                        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page
            
                    except NoSuchElementException:
                        pass
            
                    except:
                        driver.get(nxt_button[0].get_attribute('href'))
    
                time.sleep(2)
                rating_tag=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]')
                review_tag=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
                full_review_tag=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    
                for i in range(0,len(rating_tag)):
                    str=rating_tag[i].text
                    if str!='':
                        rating.append(int(str[0]))
                    else:
                        rating.append(None)
                    review.append(review_tag[i].text)
                    full_review.append(full_review_tag[i].text)
    
    except NoSuchElementException:
        pass
    
    print(len(rating),len(review))
```
