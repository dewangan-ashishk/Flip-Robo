```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementNotInteractableException
import pandas as pd
import numpy as np
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get('https://www.cardekho.com/')
```

```python
used_car_ele=driver.find_element_by_xpath('//a[@title="Cardekho Used Cars"]')
used_car_url=used_car_ele.get_attribute('href')
driver.get(used_car_url)
```

```python
for _ in range(500):
            driver.execute_script("window.scrollBy(0,20)")
city_url_ele=driver.find_elements_by_xpath('//li[@class="gsc_col-xs-4 gsc_col-sm-3 gsc_col-md-2"]//a')
city_url=[]
city_ele=driver.find_elements_by_xpath('//li[@class="gsc_col-xs-4 gsc_col-sm-3 gsc_col-md-2"]//a//span[3]')
city=[]
for ele in city_url_ele[:12]:
    city_url.append(ele.get_attribute('href'))
    
for ele in city_ele:
    city.append(ele.text)
```

```python
model=[]
brand=[]
variant=[]
price=[]
make_year=[]
fuel=[]
km=[]
engine=[]
owner=[]
transmission=[]
mileage=[]
max_power=[]
torque=[]
body_type=[]
location=[] 
```

```python
#city 11
for i in range(1):
    for j in range(6):
        driver.get(city_url[11])
        for _ in range(100):
            driver.execute_script("window.scrollBy(0,20)")
        body_type_ele=driver.find_elements_by_xpath('//ul[@class="checkBoxWrap bodyTypes cd-bodyTicon"]//li')
        b_type=body_type_ele[j].text
        try:
            driver.implicitly_wait(10)
            ActionChains(driver).move_to_element(body_type_ele[j]).click(body_type_ele[j]).perform()
            
        except ElementNotInteractableException:
            continue
        
        except StaleElementReferenceException:
            continue

        for _ in range(10000):
            driver.execute_script("window.scrollBy(0,10)")

        urls=[]
        url_ele=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7 carsName"]//a')
        for ele in url_ele:
            try:
                urls.append(ele.get_attribute('href'))
            except StaleElementReferenceException:
                continue
    
        for url in urls:
            driver.get(url)
            for _ in range(50):
                driver.execute_script("window.scrollBy(0,20)")
            try:
                price_ele=driver.find_element_by_xpath('//span[@class="amount"]')
                pr=float(price_ele.text[:-6])
                price.append(int(pr*100000))
                    
                model_ele=driver.find_element_by_xpath('//div[@class="gsc_col-xs-12"]//h1[2]')
                model.append(model_ele.text)
                    
                str=model_ele.text
                for k in range(len(str)):
                    if str[k]==' ':
                        x=k
                        break
                brand.append(str[:x])
                
                info_ele=driver.find_elements_by_xpath('//li[@class="gsc_col-xs-6"]')
                make_year.append(info_ele[0].text)
                km.append(info_ele[2].text[:-4])
                
                owner.append(info_ele[3].text)
                print(info_ele[3].text)
                fuel.append(info_ele[4].text)
                transmission.append(info_ele[5].text)
                mileage.append(info_ele[8].text[8:-5])
                
                if len(info_ele)>8:
                    engine.append(info_ele[9].text[7:-3])
                    
                    if len(info_ele)>=10:
                        max_power.append(info_ele[10].text)
                        torque.append(info_ele[11].text)
                    else:
                        max_power.append(None)
                        torque.append(None)
                    
                else:
                    engine.append(None)
                    max_power.append(None)
                    torque.append(None)
                    
            except NoSuchElementException:
                try:
                    price_ele=driver.find_element_by_xpath('//div[@class="priceSection"]//span[2]')
                    str=price_ele.text
                    price.append(str.replace(',',''))
                
                    model_ele=driver.find_element_by_xpath('//div[@class="gsc_col-xs-12"]//h1')
                    model.append(model_ele.text[5:])
                
                    str=model_ele.text[5:]
                    for k in range(len(str)):
                        if str[k]==' ':
                            x=k
                            break
                    brand.append(str[:x])
                    
                    info_ele=driver.find_elements_by_xpath('//li[@class="gsc_col-xs-6 gsc_col-md-4"]')
                    make_year.append(info_ele[0].text[10:])
                    fuel.append(info_ele[2].text[5:])
                    km.append(info_ele[3].text[11:-4])
                    engine.append(info_ele[4].text[20:-3])
                    owner.append(info_ele[5].text)
                    print(info_ele[5].text)
                
                    if len(info_ele)>=12:
                        transmission.append(info_ele[7].text[13:])
                        mileage.append(info_ele[9].text[8:-5])
                        max_power.append(info_ele[11].text)
                        torque.append(info_ele[12].text)
                    else:
                        transmission.append(None)
                        mileage.append(None)
                        max_power.append(None)
                        torque.append(None)
                except NoSuchElementException:
                    continue
    
            try:
                var_ele=driver.find_element_by_xpath('//div[@class="variant-name"]')
                variant.append(var_ele.text)
        
            except NoSuchElementException:
                variant.append(None)
   
            body_type.append(b_type)
            location.append(city[i])
            if len(owner)<len(model):
                owner.append(None)
            print(len(urls),len(owner),len(location),city[11],'\n')
```
