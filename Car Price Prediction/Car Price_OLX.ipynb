```python
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get('https://www.olx.in/')
```

```python
car_ele=driver.find_element_by_xpath('//a[@class="_2fitb"]')
car_ele.click()
```

```python
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
i=1
car_url=[]
while i<= 40:
    try:
        button = driver.find_element_by_xpath('//div[@class="JbJAl"]')
        driver.implicitly_wait(10)
        ActionChains(driver).move_to_element(button).click(button).perform()
        i+=1
    except NoSuchElementException as e:
        print("Exception Raised", e)
        break
```

```python
car_url_tag=driver.find_elements_by_xpath('//li[@class="EIR5N"]//a')
for tag in car_url_tag:
    car_url.append(tag.get_attribute('href'))
print(len(car_url))
```

```python
brand=[]
man_year=[]
model=[]
variant=[]
fuel=[]
km_driven=[]
transmission=[]
owner_count=[]
location=[]
price=[]

for url in car_url:
    driver.get(url)
    try:
        model_ele=driver.find_element_by_xpath('//div[@class="_35xN1"]')
        str=model_ele.text
        x=len(str)
        for i in range(x):
            if (str[i]==" "):
                brand.append(str[:i])
                break
        model.append(str[:(x-6)])
        man_year.append(str[-5:-1])
    except NoSuchElementException:
        continue
    
    try:
        variant_ele=driver.find_element_by_xpath('//div[@class="_3tLee"]')
        variant.append(variant_ele.text)
    except NoSuchElementException:
        variant.append(None)
    
    try:
        info_ele=driver.find_elements_by_xpath('//div[@class="_3qDp0"]')
        fuel.append(info_ele[0].text)
        str=info_ele[1].text
        x=len(str)
        km_driven.append(str[:(x-3)])
        transmission.append(info_ele[2].text)
    except NoSuchElementException:
        fuel.append(None)
        km_driven.append(None)
        transmission.append(None)
    except IndexError:
        fuel.append(None)
        km_driven.append(None)
        transmission.append(None)

    try:
        owner_info_ele=driver.find_elements_by_xpath('//div[@class="_1gasz"]')
        str=owner_info_ele[0].text
        owner_count.append(str[0])
        address=owner_info_ele[1].text
        for i in range(len(address)):
            if address[i]==' ':
                x=i
        location.append(address[x+1:])
    except NoSuchElementException:
        owner_count.append(None)
        location.append(None)

    try:
        price_ele=driver.find_element_by_xpath('//div[@class="_3FkyT"]')
        str=price_ele.text
        price.append(str[2:])
    except NoSuchElementException:
        price.append(None)
    
olx=pd.DataFrame({'Brand':(brand),
                  'Model':(model),
                  'Manufacturing Year':(man_year),
                 'Variant':(variant),
                 'Fuel':(fuel),
                 'KM Driven':(km_driven),
                 'Transmission':(transmission),
                 'Number of Owners':(owner_count),
                 'Location':(location),
                 'Price':(price)})
olx
```
