# 1.  Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
tag=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]//td')
```

```python
rank=[]
name=[]
artist=[]
views=[]
upload_date=[]

for i in range(0,30):
    rank.append(tag[6*i].text)
    str=tag[6*i+1].text
    name.append(str[:-4])
    artist.append(tag[6*i+2].text)
    views.append(tag[6*i+3].text)
    upload_date.append(tag[6*i+4].text)
Data=pd.DataFrame({'Rank':(rank),
                  'Name':(name),
                  'Artist':(artist),
                  'Views (billions)':(views),
                  'Upload Date':(upload_date)})
Data
```

# 2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/.

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.bcci.tv')
```

```python
international=driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]')
international.click()
fixtures=driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]')
fixtures.click()
```

```python
title_tag=driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
series_tag=driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]')
location=driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]//span')
date_tag=driver.find_elements_by_xpath('//span[@class="fixture__date"]')
month_tag=driver.find_elements_by_xpath('//span[@class="fixture__month"]')
time_tag=driver.find_elements_by_xpath('//span[@class="fixture__time"]')
```

# 5.  Scrape the details of trending repositories on Github.com. Url = https://github.com/

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
driver=webdriver.Chrome('chromedriver.exe')
driver.get(' https://github.com')
```

```python
explore=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary')
explore.click()
```

```python
trending=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a')
driver.get(trending.get_attribute('href'))
```

```python
repo=driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]//a')
```

```python
title=[]
link=[]
for i in  range(0,len(repo)):
    str=repo[i].text
    x=str.find('/')
    title.append(str[x+2:])
    link.append(repo[i].get_attribute('href'))
```

```python
description=[]
contributor=[]
language=[]
for i in link:
    driver.get(i)
    time.sleep(2)
    
    try:
        desc=driver.find_element_by_xpath('//p[@class="f4 mt-3"]')
        description.append(desc.text)
    except NoSuchElementException:
        description.append('-')
        
    lang=driver.find_elements_by_xpath('//a[@class="d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"]//span')
    str=''
    for j in range(0,len(lang)):
        if j%2==0:
            str=str+lang[j].text+','

    language.append(str)
    
    try:
        cont=driver.find_elements_by_xpath('//h2[@class="h4 mb-3"]//span')
        x=len(cont)
        if x==0:
            contributor.append('0')
        else:
            contributor.append(cont[-1].text)
    except NoSuchElementException:
        contributor.append('-')
```

```python
Data=pd.DataFrame({'Repository Title':(title),
                  'Repository Description':(description),
                  'Contributor Count':(contributor),
                  'Language Used':(language)})
Data
```

# 7.  Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.naukri.com')
```

```python
recruiter=driver.find_element_by_xpath('//a[@title="Search Recruiters"]')
link=recruiter.get_attribute('href')
driver.get(link)
```

```python
search=driver.find_element_by_xpath('//input[@class="sugInp"]')
search.send_keys("Data Science")
```

```python
drop_down=driver.find_element_by_xpath('//div[@class="Sbtn"]')
drop_down.click()
```

```python
search_button=driver.find_element_by_xpath('//button[@class="fl qsbSrch blueBtn"]')
search_button.click()
```

```python
info=driver.find_elements_by_xpath('//p[@class="highlightable"]')
skills_ele=driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]')
```

```python
name=[]
company=[]
designation=[]
skills=[]
location=[]

for i in range(0,len(skills_ele)):
    str=info[i].text
    x=str.find('\n')
    name.append(str[:x])
    str=str.replace(str[:x+1],' ')
    x=str.find('\n')
    designation.append(str[0:x])
    str=str.replace(str[0:x+1],' ')
    x=str.find('\n')
    if x==-1:
        company.append(str)
        location.append('-')
    else:
        company.append(str[0:x])
        str=str.replace(str[0:x+1],' ')
        location.append(str[1:])
    skills.append(skills_ele[i].text)
        
Data=pd.DataFrame({'Name':(name),
                  'Designation':(designation),
                  'Company':(company),
                  'Skills':(skills),
                  'Location':(location)})
Data
```

# 9. Scrape the details most watched tv series of all time from imdb.com. Url =  https://www.imdb.com/list/ls095964455/

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://www.imdb.com/list/ls095964455')
```

```python
rank_ele=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//span[1]')
name_ele=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//a')
year_ele=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]//span[2]')
genre_ele=driver.find_elements_by_xpath('//p[@class="text-muted text-small"]//span[5]')
runtime_ele=driver.find_elements_by_xpath('//p[@class="text-muted text-small"]//span[3]')
rating_ele=driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]//span[2]')
votes_ele=driver.find_elements_by_xpath('//p[@class="text-muted text-small"]//span[2]')
```

```python
rank=[]
name=[]
year=[]
genre=[]
runtime=[]
rating=[]
votes=[]

for i in range(100):
    rank.append(rank_ele[i].text)
    name.append(name_ele[i].text)
    year.append(year_ele[i].text)
    genre.append(genre_ele[i].text)
    runtime.append(runtime_ele[i].text)
    rating.append(rating_ele[i].text)
    votes.append(votes_ele[2*i+1].text)
    
Data=pd.DataFrame({'Rank':(rank),
                  'Name':(name),
                 'Year Span':(year),
                 'Genre':(genre),
                  'Runtime':(runtime),
                  'Ratings':(rating),
                  'Votes':(votes)})

Data
```

```python
info=driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td')
```

```python
del info[0:7]
```

```python
dataset=[]
datatype=[]
task=[]
attributetypes=[]
instances=[]
attributes=[]
year=[]
for i in range(588):
    dataset.append(info[7*i].text)
    datatype.append(info[7*i+1].text)
    task.append(info[7*i+2].text)
    attributetypes.append(info[7*i+3].text)
    instances.append(info[7*i+4].text)
    attributes.append(info[7*i+5].text)
    year.append(info[7*i+6].text)
    
Data=pd.DataFrame({'Data Set Name':(dataset),
                  'Data Types':(datatype),
                  'Default Task':(task),
                  'Attribute Types':(attributetypes),
                  'No. of Instances':(instances),
                  'No. of Attributes':(attributes),
                  'Year':(year)})

Data
```
