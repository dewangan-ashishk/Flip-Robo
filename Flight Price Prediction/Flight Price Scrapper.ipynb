```python
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
import time
import warnings
warnings.filterwarnings('ignore')
driver=webdriver.Chrome('/Users/ashish/Desktop/Python/Flip Robo/Flight Price Prediction/chromedriver-3')
```

```python
#we will find all the links and then use the link for flight details page and edit it to get the flight price for our desired dates and location
links_all=driver.find_elements_by_xpath('.//a')
for a in links_all:
    print(a.get_attribute('href'))

link=links_all[-1].get_attribute('href')
```

```python
#date range
dates=pd.date_range('11/03/2021', periods=90)
dates
```

```python
city=['DEL','BOM','BLR','HYD',
      'CCU','MAA','AMD','GOI']
c=len(city)
```

```python
airline=[]
date_enq=[]
date_journey=[]
source=[]
departure=[]
arrival=[]
destination=[]
duration=[]
stops=[]
route=[]
price=[]
```

```python
links=[]
link=link.replace('DEL','SRC')
link=link.replace('BLR','DST')
for i in range(c):
    for j in range(c):
        if city[i]==city[j]:
            continue
        else:
            for dt in dates:
                new_link=link.replace('SRC',city[i])
                new_link=new_link.replace('DST',city[j])
                new_link=new_link.replace('01/11/2021',dt.strftime("%d/%m/%Y"))
                links.append(new_link)
```

```python
for path in links:
    driver.get(path)
    
    time.sleep(5)
    
    for _ in range(500):
            driver.execute_script("window.scrollBy(0,20)")
            
    #extracting flight carrier information
    airline_ele=driver.find_elements_by_xpath('//div[@class="makeFlex align-items-center  "]')

    #extracting time, city and price information
    time_ele=driver.find_elements_by_xpath('//div[@class="makeFlex flex-column flightTimeInfo"]')
    duration_ele=driver.find_elements_by_xpath('//div[@class="stop-info flexOne"]//p')

    #extracting price
    price_ele=driver.find_elements_by_xpath('//div[@class="textRight flexOne"]')

    for x in range(len(airline_ele)):
        if airline_ele[x].text=='':
            continue
        else:
            airline.append(airline_ele[x].text)
    
        date_enq.append('03/11/2021')
        date_journey.append(path[59:59+8])
    
        if price_ele[x].text=='':
            price.append(None)
        else:
            price.append(int(price_ele[x].text[2:].replace(',','')))
        
        source.append(time_ele[2*x].text[6:])
        departure.append(time_ele[2*x].text[:5])
        destination.append(time_ele[2*x+1].text[6:])
        arrival.append(time_ele[2*x+1].text[:5])
    
        if duration_ele[3*x].text=='':
            duration.append(None)
            stops.append(None)
            route.append(None)
        else:
            dura=int(duration_ele[3*x].text[:2]) + (int(duration_ele[3*x].text[-4:-2]))/60
            duration.append(dura)
        
            if duration_ele[3*x+2].text[0]=='N':
                stops.append('0')
                route.append(None)
            else:
                stops.append(duration_ele[3*x+2].text[0])
                route.append(duration_ele[3*x+2].text[11:])
    print(path[59:59+8])
```

```python
import pandas as pd
data=pd.DataFrame({'Airline': (airline),
                   'Date_enq': (date_enq),
                   'Date_journey': (date_journey), 
                 'Source': (source),
                 'Departure': (departure),
                  'Destination': (destination),
                  'Arrival': (arrival),
                  'Duration': (duration),
                  'Stops': (stops),
                  'Route': (route),
                  'Price': (price)})
data
```

```python
data.to_csv('Flight_Price.csv')
```
